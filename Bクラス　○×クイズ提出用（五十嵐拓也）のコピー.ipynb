```python
from IPython.display import display
import pandas as pd
a = 'a'
b = 'b'
c = 'c'
d = 'd'
e = 'e'
f = 'f'
g = 'g'
h = 'h'
i = 'i'

count = 0
used = []
display(pd.DataFrame([[a,b,c],[d,e,f],[g,h,i]]))

while True:
    if count %2 == 0:
        sign = '○'
        player = '1'
    else:
        sign = '×'
        player = '2'
        
    while True:
        p1 = input('プレーヤー'+player+ 'の番です：')
        if p1 not in ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']:
            print('正しい文字を入力してください')
        elif p1 not in used:
            break
        else:
            print('既に埋まっています')
    
    used.append(p1)
    count += 1
    
    if p1 == a:
        a = sign
    elif p1 == b:
        b = sign
    elif p1 == c:
        c = sign
    elif p1 == d:
        d = sign
    elif p1 == e:
        e = sign
    elif p1 == f:
        f = sign
    elif p1 == g:
        g = sign
    elif p1 == h:
        h = sign
    elif p1 == i:
        i = sign
    
    display(pd.DataFrame([[a,b,c],[d,e,f],[g,h,i]]))
    if a == b == c:
        print('プレーヤー'+ player+ 'の勝ちです')
        break
    elif d == e == f:
        print('プレーヤー'+ player+ 'の勝ちです')
        break
    elif g == h == i:
        print('プレーヤー'+ player+ 'の勝ちです')
        break
    elif a == d == g:
        print('プレーヤー'+ player+ 'の勝ちです')
        break
    elif b == e == h:
        print('プレーヤー'+ player+ 'の勝ちです')
        break
    elif c == f == i:
        print('プレーヤー'+ player+ 'の勝ちです')
        break
    elif a == e == i:
        print('プレーヤー'+ player+ 'の勝ちです')
        break
    elif c == e == g:
        print('プレーヤー'+ player+ 'の勝ちです')
        break
    elif count == 9:
        print('ゲーム終了、引き分けです')
        break
```
